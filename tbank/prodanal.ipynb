```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('table.xlsx')
```

```python
df
```

```python
df_tr = df[:56]
```

```python
fig, ax = plt.subplots(1,1)
for col in ['Регистрации',	'Открыли поиск',	'Просмотрели авто',	'Забронировали',	'Первая поездка']:
    sns.lineplot(x = df_tr.index, y = df_tr[col]/df_tr['Установки'], label = col, ax = ax)
ax.set_xlabel('День')
ax.set_ylabel('Доля от установок')
sns.move_legend(ax, "upper left", bbox_to_anchor=(1, 1))
```
