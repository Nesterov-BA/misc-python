```python
import numpy as np
import pandas as pd
```

```python
def part(arr):
    pivot = arr[(len(arr))//2]
    print('pivot = ',(len(arr))//2)
    larr = []
    rarr = []
    for i in range(len(arr)):
        if arr[i] < pivot:
            larr.append(arr[i])
        elif arr[i] > pivot:
            rarr.append(arr[i])
    index = len(larr)
    return larr, rarr, index
```

```python
part([1,3,2,10,5])
```

```python
def kth_element(arr, k):
    while True:
        print(len(arr))
        larr, rarr, ind = part(arr)
        print(larr, rarr)
        if ind == k:
            return larr[ind-1]
        elif ind > k:
            arr = larr
        else: 
            arr = rarr
```
